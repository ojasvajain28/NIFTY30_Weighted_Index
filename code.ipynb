```python
ret_close_price = close_price.pct_change()
```

```python
ret_close_price
```

```python
describe = ret_close_price.describe()
```

```python
std_stocks = describe.loc['std',:]
std_stocks
```

```python
top_30=std_stocks.sort_values()[:30]
```

```python
top_30 = pd.DataFrame(top_30)
```

```python
top_30['reci_vol'] = 1/top_30['std']
top_30
```

```python
s = top_30['reci_vol'].sum()
s
```

```python
top_30['weight'] = top_30['reci_vol']/s*100
top_30
```

```python
f = plt.figure()
f.set_figwidth(40)
f.set_figheight(10)
plt.plot(top_30['weight'], linewidth=3, marker='o', )
```
